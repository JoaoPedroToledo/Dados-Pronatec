```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('PDA_UNIDADES_RF_EPCT_CSV.csv', sep=';', encoding='utf8')
```

```python
#Primeiramente temos que ler a tabela do computador usando o metodo read_csv e passando o caminho da tabela salvo.
#Ao atribuir o metodo a variavel 'df', basta utilizar a variavel pra outros metodos na tabela.
#Ao executar a variavel "df", é mostrado toda a tabela

df
```

```python
# Mostra quantas unidades existem no arquivo
df.count()
```

```python
# Mostra quantas unidades existem no arquivo agrupadas por estado, nome da mesorregião e nome dos municípios
df.groupby('NOME_MESORREGIAO_UNIDADE')['NOME_MUNICIPIO_UNIDADE'].value_counts()
```

```python
# Mostra quantas unidades existem no arquivo agrupadas por região e estado
df.groupby('NOME_REGIAO_UNIDADE')['SIGLA_UF_UNIDADE'].value_counts()
```

```python
# Mostra quantas unidades existem no arquivo agrupadas por estado, nome da microrregião e nome dos municípios
df.groupby('NOME_MICRORREGIAO_UNIDADE')['NOME_MUNICIPIO_UNIDADE'].value_counts()
```

```python
df['SIGLA_UF_UNIDADE'].value_counts().plot(kind ='bar', figsize = (11,5), grid = False, color = 'powderblue')
plt.title('Quantidade de unidades por estado')
```

```python
#O gráfico acima mostra as unidades federais por estado, nela podemos analisar que
#em Minas Gerais está localizado grande parte dos institutos federais
```
